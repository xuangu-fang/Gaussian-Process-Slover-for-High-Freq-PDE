```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# possion-2d-sinsin (100k epochs)
reso = np.array([10,50, 100,200,300,400])
tesr_err_SM = np.array([0.999, 0.6261,  0.3760,  0.001, 0.0003, 0.000065])
tesr_err_SMM = np.array([0.999, 0.726, 0.1697, 0.0336, 0.0002, 0.000068])


plt.plot(reso,np.log10( tesr_err_SM), marker='s', label='GP-HM-GM',markersize=10)
plt.plot(reso,np.log10( tesr_err_SMM),marker='o', label='GP-HM-StM',markersize=10)


# plt.plot(reso, tesr_err_SM, 'r', label='SM')
# plt.plot(reso,( tesr_err_SMM), 'b', label='SMM')

x_ticks = reso
plt.xticks(x_ticks,['10x10','50x50','100x100','200x200','300x300','400x400'], fontsize=10)
plt.xlabel('Resolution',fontsize=20)
plt.ylabel('${Log}_{10}$ (Relative $L_2$ Error)',fontsize=20)
plt.legend(fontsize=20)

plt.savefig('../figs/resolution/poission-2d-sinsin.pdf',bbox_inches='tight')
```
